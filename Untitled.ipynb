```python
import pandas as pd

# Load training data
train_data = pd.read_csv('training_set.csv')
```

```python
# Check for missing values
missing_values = train_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

# List of continuous and categorical features
con_features = train_data.select_dtypes(include=[np.number]).columns.tolist()
con_features.remove('SalePrice')
cat_features = train_data.select_dtypes(exclude=[np.number]).columns.tolist()

# Impute missing values
con_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='constant', fill_value='Not_Avail')

train_data[con_features] = con_imputer.fit_transform(train_data[con_features])
train_data[cat_features] = cat_imputer.fit_transform(train_data[cat_features])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Univariate Analysis for SalePrice
plt.figure(figsize=(10, 6))
sns.histplot(train_data['SalePrice'], kde=True)
plt.title('Distribution of SalePrice')
plt.show()

# Univariate Analysis for Numerical Features
train_data[con_features].hist(bins=20, figsize=(20, 15))
plt.show()

# Calculate grid size
num_features = len(con_features)
num_cols = 4
num_rows = math.ceil(num_features / num_cols)

# Bivariate Analysis for SalePrice vs other numerical features
plt.figure(figsize=(20, 15))
for i, feature in enumerate(con_features):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.scatterplot(x=train_data[feature], y=train_data['SalePrice'])
    plt.title(f'SalePrice vs {feature}')
plt.tight_layout()
plt.show()

# Correlation Matrix for Numeric Features Only
numeric_features = train_data.select_dtypes(include=[np.number]).columns.tolist()
plt.figure(figsize=(12, 8))
corr_matrix = train_data[numeric_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Ordinal Encoding for categorical features
ordinal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Not_Avail')),
    ('encoder', OrdinalEncoder())
])

# OneHotEncoding for final pipeline
final_pipeline = ColumnTransformer(transformers=[
    ('num', con_imputer, con_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
])
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Feature selection with Recursive Feature Elimination (RFE)
X = train_data.drop(columns=['SalePrice'])
y = train_data['SalePrice']

# Apply final pipeline for feature transformation
X_transformed = final_pipeline.fit_transform(X)

rfe = RFE(estimator=LinearRegression(), n_features_to_select=10)
X_rfe = rfe.fit_transform(X_transformed, y)

# Get selected feature names (requires handling ColumnTransformer output)
feature_names = (con_features + list(final_pipeline.named_transformers_['cat'].get_feature_names_out(cat_features)))
selected_features = [feature_names[i] for i in range(len(feature_names)) if rfe.support_[i]]
print("Selected features:\n", selected_features)
```

```python
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Ridge, Lasso

# Ridge and Lasso models
ridge_model = Ridge()
lasso_model = Lasso()
```

```python
from sklearn.model_selection import GridSearchCV

# parameter grid
param_grid = {'alpha': [0.1, 1.0, 10.0, 100.0]}

# GridSearchCV for Ridge
ridge_grid = GridSearchCV(ridge_model, param_grid, cv=5, scoring='neg_mean_squared_error')
ridge_grid.fit(X_train, y_train)

# GridSearchCV for Lasso
lasso_grid = GridSearchCV(lasso_model, param_grid, cv=5, scoring='neg_mean_squared_error')
lasso_grid.fit(X_train, y_train)
```

```python
# Predictions for Ridge
ridge_pred_train = ridge_grid.predict(X_train)
ridge_pred_test = ridge_grid.predict(X_test)

# Predictions for Lasso
lasso_pred_train = lasso_grid.predict(X_train)
lasso_pred_test = lasso_grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Evaluation metrics for Ridge
ridge_mse_train = mean_squared_error(y_train, ridge_pred_train)
ridge_rmse_train = np.sqrt(ridge_mse_train)
ridge_mae_train = mean_absolute_error(y_train, ridge_pred_train)
ridge_r2_train = r2_score(y_train, ridge_pred_train)

ridge_mse_test = mean_squared_error(y_test, ridge_pred_test)
ridge_rmse_test = np.sqrt(ridge_mse_test)
ridge_mae_test = mean_absolute_error(y_test, ridge_pred_test)
ridge_r2_test = r2_score(y_test, ridge_pred_test)

print("Ridge Regression Evaluation Metrics:")
print(f"Train MSE: {ridge_mse_train}, RMSE: {ridge_rmse_train}, MAE: {ridge_mae_train}, R²: {ridge_r2_train}")
print(f"Test MSE: {ridge_mse_test}, RMSE: {ridge_rmse_test}, MAE: {ridge_mae_test}, R²: {ridge_r2_test}")

# Evaluation metrics for Lasso
lasso_mse_train = mean_squared_error(y_train, lasso_pred_train)
lasso_rmse_train = np.sqrt(lasso_mse_train)
lasso_mae_train = mean_absolute_error(y_train, lasso_pred_train)
lasso_r2_train = r2_score(y_train, lasso_pred_train)

lasso_mse_test = mean_squared_error(y_test, lasso_pred_test)
lasso_rmse_test = np.sqrt(lasso_mse_test)
lasso_mae_test = mean_absolute_error(y_test, lasso_pred_test)
lasso_r2_test = r2_score(y_test, lasso_pred_test)

print("Lasso Regression Evaluation Metrics:")
print(f"Train MSE: {lasso_mse_train}, RMSE: {lasso_rmse_train}, MAE: {lasso_mae_train}, R²: {lasso_r2_train}")
print(f"Test MSE: {lasso_mse_test}, RMSE: {lasso_rmse_test}, MAE: {lasso_mae_test}, R²: {lasso_r2_test}")
```

```python
ridge_coef = ridge_grid.best_estimator_.coef_
lasso_coef = lasso_grid.best_estimator_.coef_

def plot_feature_importance(coefs, feature_names, model_name):
    coef_df = pd.DataFrame({'Feature': feature_names, 'Importance': coefs})
    coef_df = coef_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=coef_df)
    plt.title(f'Feature Importance for {model_name}')
    plt.show()

plot_feature_importance(ridge_coef, selected_features, 'Ridge Regression')
plot_feature_importance(lasso_coef, selected_features, 'Lasso Regression')
```
